```python
from sklearn.datasets import fetch_openml
digits = fetch_openml(name='mnist_784', version=1)
plt.imshow(digits.data[0].reshape(28,28), cmap=plt.cm.gray_r)
```

```python
from sklearn.model_selection import train_test_split
import numpy as np 

scaled_data = digits.data / 255

# ターゲットをone_hot表現に変換
one_hot_target = np.zeros((digits.target.shape[0], 10))
for i in range(digits.target.shape[0]):
    one_hot_target[i, int(digits.target[i])] = 1

X_train, X_test, y_train, y_test = train_test_split(scaled_data, one_hot_target)
```
