```python
# 関数の描画
import numpy as np

def function(x):
    return x[0]**2 + x[1]**2
```

```python
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D 

fig = plt.figure()
ax = fig.gca(projection='3d')

x = np.arange(-2, 2, 0.05) 
y = np.arange(-2, 2, 0.05)  
xx, yy = np.meshgrid(x, y)

zz = function([xx, yy])

ax.plot_surface(xx, yy, zz, rstride=1, cstride=1, cmap='hsv', linewidth=0.3)
```
